```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Suchana Hazra\OneDrive\Desktop\ml projects\temp-icecreamSales.csv")
df
```

```python
df['profits'] = df.iloc[:,0].values
df
```

```python
df.drop(['Ice Cream Profits'],axis = 'columns',inplace = True)
df
```

```python
%matplotlib inline
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')
plt.scatter(df.Temperature,df.profits,color='red',marker='+')
```

```python
new_df = df.drop(['profits'],axis = 'columns')
new_df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Temperature']],df.profits)
```

```python
%matplotlib inline
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')
plt.scatter(df.Temperature,df.profits,color='red',marker='+')
plt.plot(df.Temperature,reg.predict(df[['Temperature']]),color = 'blue')
```
